```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dots = np.linspace(0,1,10)
dots
```

```python
noise = (np.random.random(10)*2-1)*0.1
noise
```

```python
experimental_data = np.sin(2*np.pi*dots)+noise
experimental_data
```

```python
plt.plot(dots, experimental_data)
```

```python
from scipy.interpolate import interp1d
```

```python
dots_interp = np.linspace(0,1,100)
dots_interp
```

```python
linear_interp = interp1d(dots, experimental_data)
linear_interp
```

```python
linear_results = linear_interp(dots_interp)
linear_results
```

```python
plt.figure(figsize=(15,8))
plt.plot(dots, experimental_data)
plt.plot(dots_interp, linear_results)
```

```python
cubic_interp = interp1d(dots, experimental_data, kind = "cubic")
cubic_interp
```

```python
cubic_results = cubic_interp(dots_interp)
```

```python
plt.figure(figsize=(15,8))
plt.plot(dots, experimental_data)
plt.plot(dots_interp, cubic_results)
```

```python
x_data = np.linspace(-5,5,50)
y_data = 2.9*np.sin(1.5*x_data)+np.random.normal(size=50)
```

```python
np.random.normal(size=50)
```

```python
plt.plot(x_data, y_data, "ro")
```

```python
def func(x, a, b):
    return a*np.sin(b*x)
```

```python
from scipy.optimize import curve_fit
```

```python
params, _ = curve_fit(func, x_data, y_data, p0=[2,2])
print(params)
```

```python
y_data_fitted = func(x_data, 2.62291091, 1.44754323)
```

```python
plt.plot(x_data, y_data, "ro")
plt.plot(x_data, y_data_fitted)
```
