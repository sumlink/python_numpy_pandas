```python
import pandas as pd
```

```python
cities_capital = pd.DataFrame({
    "countries":["Russia", "USA", "UK"],
    "cities":["Moscow", "Los Angeles", "London"],
    "capital":[1,0,1]
})
cities_capital
```

```python
cities_capital_new = pd.DataFrame({
    "countries":["Russia", "Spain", "Germany"],
    "cities":["Ekaterinburg", "Barcelona", "Munich"],
    "capital":[0,0,0]
})
cities_capital_new
```

```python
all_cities_capital = pd.concat([cities_capital, cities_capital_new], axis = 0)
all_cities_capital
```

```python
all_cities_capital.index = range(0,6)
```

```python
all_cities_capital
```

```python
population = pd.DataFrame({
    "population":[20, 5, 10,1,3,5],
    "language":["Russian", "English", "English", "Russian", "Spanish", "German"]
})
```

```python
all_cities_capital = pd.concat([all_cities_capital, population], axis = 1)
all_cities_capital
```

```python
continents = pd.DataFrame({
    "continent": ["Europe", "North America", "Asia"], 
    "countries": ["Russia", "USA", "China"]
})
continents
```

```python
pd.merge(all_cities_capital, continents, left_on="countries", right_on="countries", how="left")
```
