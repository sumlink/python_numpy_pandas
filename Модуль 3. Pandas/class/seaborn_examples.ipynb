```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./data/Pokemon.csv", index_col = 0)
df
```

```python
sns.lmplot(x="Attack", y="Defense", data=df, fit_reg=False, hue="Stage")
```

```python
sns.boxplot(data=df)
```

```python
box_df = df.drop(["Total", "Stage", "Legendary"], axis=1)
box_df
```

```python
sns.boxplot(data=box_df)
```

```python
plt.figure(figsize=(15,8))
sns.violinplot(x = "Type 1", y = "Attack", data=df)
```

```python
sns.swarmplot(x = "Type 1", y = "Attack", data=df)
```

```python
plt.figure(figsize=(15,8))
sns.violinplot(x = "Type 1", y = "Attack", data=df, inner = None)
sns.swarmplot(x = "Type 1", y = "Attack", data=df, color = "k")
```
